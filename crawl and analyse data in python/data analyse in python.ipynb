```python
man_ad1=man[["id-name","height","weight","address"]]
woman_ad1 =woman[["id-name","height","weight","address"]]
man_ad1["bmi"]=man_ad1["weight"]/(man_ad1["height"]**2)*10000
woman_ad1["bmi"]=woman_ad1["weight"]/(woman_ad1["height"]**2)*10000
```

```python
man_ad1 =man_ad1[man_ad1["bmi"]<23.9]
man_ad1 =man_ad1[man_ad1["bmi"]>18.5]
woman_ad1 =woman_ad1[woman_ad1["bmi"]<23.9]
woman_ad1 =woman_ad1[woman_ad1["bmi"]>18.5]
```

```python
# 9.查看薪资排名前10 的省份
# 薪资为保密,0,空值的不进行计算
# 薪资在区间内的取其均值
# 50000以上的视为50000  2000以下的视为2000
human1=human[human["salary"]!=0]
human2=human1[human1["salary"]!="保密"]
human3=human2[~human2["salary"].isnull()]
human4=human3.replace({"5000 以上":"5000","50000 以上":"50000","2000 以下":"2000"})
```

```python
def s6(x):
    x = x.split("~")
    sum1 =0
    for i in x:
        sum1 +=int(i)
    return sum1/len(x)
human4["salary"]=human4["salary"].apply(s6)
```

```python
salary = human4.groupby("address")["salary"].mean()
salary_mean = salary.sort_values(ascending = False).head(10)
salary_mean
```
